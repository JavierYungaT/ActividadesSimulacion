```python
#Dibujo de distribución binomial
 # Importar paquetes requeridos
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib.style as style
from IPython.core.display import HTML

 #PLOTTING CONFIG Configuración de dibujo
%matplotlib inline

style.use('fivethirtyeight')
plt.rcParams['figure.figsize']=(14,7)
plt.figure(dpi=100)

 #PMF Trazado de la función de calidad de probabilidad
plt.bar(left=(np.arange(20)), height=(stats.binom.pmf(np.arange(20), p = 0.5, n = 20)), width = 0.75, alpha = 0.75) # binom.pmf es la masa de probabilidad del binomio
 #n = 20, P = 0.5, dibuje como un gráfico de barras

#CDF
plt.plot (np.arange (20), stats.binom.cdf (np.arange (20), p = 0.5, n = 20), color = "# fc4f30") # Dibuje la función de densidad acumulativa del binomio Curva

 #LEGEND Legend
plt.text(x=7.5,y=0.2,s="pmf(binomed)",alpha=0.75,weight="bold",color="#008fd5")
plt.text(x=14.5,y=0.9,s="cdf",rotation=.75,weight="bold",color="#fc4f30")
```
