# ¿Qué es Tausworthe? 

Los generadores de Tausworthe producen números pseudo-aleatorios con un período grande.

```python
#Algoritmo Tausworthe 
from itertools import zip_longest
# Ingresamos por medio de consola los valores a requerir 

x = int(input("Ingrese el valor de x: "))
y = int(input("Ingrese el valor de y: "))

#Generamos la operacón XOR
def operacionXOR(a,b):
  ab = 1
  if a  == b:
    ab = 0
  return ab

bits = []
b = (2**y) - 1

for i in range(0,b):
  bits.append(0)


for i in range(0,y):
  bits[i] = 1
  bits.append(1)

a = y+1
for i in range(a,len(bits)):
  i1 = i - x
  i2 = i - y
  bits[i] = operacionXOR(bits[i1],bits[i2])


def binaryToDecimal(binario):
  a = 0
  p = 3
  for i in range(0,len(binario)):
    if binario[i] == 1:
      a += 2**(p-i)
  return a



test_list = bits


def group_elements(n, iterable, padvalue='1'):
    return zip_longest(*[iter(iterable)]*n, fillvalue=padvalue)

print("\n","It.", "\t", "Base 2", "\t","Base 10","\t","Ui","\n")
d = 0
for output in group_elements(4,test_list):
    lst_new = [str(a) for a in output]
    print(d,"\t" ," ".join(lst_new), "\t", binaryToDecimal(output), "\t", "\t", binaryToDecimal(output)/16)
    d +=1
```
